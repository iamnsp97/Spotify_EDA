```python
df_spfy["duration"] = df_spfy['duration_ms'].apply( lambda x : round(x/1000))
df_spfy.drop('duration_ms' , inplace = True , axis = 1)
df_spfy
```

```python
corr_df = df_spfy.drop(['key','mode', 'explicit'], axis = 1).corr(method= "pearson")
plt.figure(figsize= (16, 7))
heatmap = sns.heatmap(corr_df , annot = True, fmt =".1g", vmin = -1 , vmax = 1, center = 0 , cmap = "magma", linewidths = 1, linecolor = "Black")
heatmap.set_title( " Correlation HeatMap Between variable ")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=360)
```

```python
df_genre = pd.read_csv('D:/4/Datasets/SpotifyFeatures.csv')
```

```python
df_genre.head(5)
```

```python
df_genre.info()
```

```python
df_genre['duration'] = df_genre['duration_ms'].apply(lambda x: round(x/1000))
df_genre.drop('duration_ms', inplace =True, axis = 1)
```

```python
plt.title("Duration of Songs in different genres")
sns.color_palette("rocket_r", as_cmap= False)
sns.barplot(y="genre", x ="duration", data=df_genre)
plt.xlabel("duration in mill seconds")
#plt.xticks(rotation= 90)
plt.ylabel("Genre")
```

```python
plt.title("Duration of Songs in different genres")
sns.color_palette("rocket_r", as_cmap= False)
plt.figure(figsize=(13,6))
sns.lineplot(x="genre", y ="duration", data=df_genre)
plt.ylabel("duration in mill seconds")
plt.xticks(rotation= 90)
plt.xlabel("Genre")
```

```python
genre =df_genre.genre
total_dr = df_genre.duration 
fig_dim = (13, 6)
fig, ax = plt.subplots(figsize = fig_dim)
sns.set_style(style="whitegrid")
fig =sns.lineplot(x=genre, y = total_dr ,ax = ax).set(title="Line plot for duration vs years")
plt.xticks(rotation = 90)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(13,6))
famous= df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(y="genre", x ="popularity", data=famous).set(title="Popularity vs genre")
plt.xlabel("Popularity")
plt.ylabel("genre")
```
